```python
import pandas as pd
```

```python
df = pd.read_csv("holerite.csv", sep=";")
df.columns
```

```python
df[['Mês', 'Ano']] = df['Mês/Ano'].str.split('-', expand=True)
df.drop('Mês/Ano', axis=1, inplace=True)
df
```

```python
nova_ordem = [
    'Nome Completo', 'Email', 'Mês', 'Ano', 'Cargo', 'CBO', 'PIS',
    'Salário Base', 'Comissão', 'Valores Adicionais', 'Bonificações',
    'Total Vencimentos', 'INSS (%)', 'INSS (R$)', 'IRRF (%)', 'IRRF (R$)',
    'Plano de Saúde', 'Total Descontos', 'Líquido a Receber',
    'Carga Horária', 'Horas Extras', 'Base FGTS', 'FGTS do Mês',
    'Base INSS', 'Base IRRF'
]

df = df[nova_ordem]
df
```

```python
df.loc[df['Mês'] == "Jan", 'Mês'] = "Janeiro"
df.loc[df['Mês'] == "Feb", 'Mês'] = "Fevereiro"
df.loc[df['Mês'] == "Mar", 'Mês'] = "Março"
df.loc[df['Mês'] == "Apr", 'Mês'] = "Abril"
df.loc[df['Mês'] == "May", 'Mês'] = "Maio"
df.loc[df['Mês'] == "Jun", 'Mês'] = "Junho"
df.loc[df['Mês'] == "Jul", 'Mês'] = "Julho"
df.loc[df['Mês'] == "Aug", 'Mês'] = "Agosto"
df.loc[df['Mês'] == "Sep", 'Mês'] = "Setembro"
df.loc[df['Mês'] == "Oct", 'Mês'] = "Outubro"
df.loc[df['Mês'] == "Nov", 'Mês'] = "Novembro"
df.loc[df['Mês'] == "Dec", 'Mês'] = "Dezembro"


df.to_csv("novoHolerite.csv", sep=";", index=False, encoding="utf-8")

```
